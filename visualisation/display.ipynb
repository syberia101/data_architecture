```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import scipy.stats as stats
```

```python

pd.set_option("display.max_columns", None)
# Puts the scrollbar next to the DataFrame
```

```python
df = pd.read_csv('/Users/derib/data_architecture/visualisation/CO2_Emissions_1960-2018.csv')
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.plot(kind='box',figsize=(20, 10)
)

plt.ylabel('CO2 emissions (metric tons)')
plt.xlabel('Years')
plt.xticks(rotation='vertical')

plt.show()
```

### Some data transformation to get the countries with the bigger gap between 1960 and 2018

```python
df['Gap'] = df.loc[:, '1960':'2018'].max(axis=1) - df.loc[:, '1960':'2018'].min(axis=1)
df_sorted = df.sort_values(by='Gap', ascending=False)
df_smaller = df_sorted.head(20)
df_smaller
```

```python
df_smaller.drop(columns=["Gap"]).plot(kind='box',figsize=(20, 10)
)

plt.ylabel('CO2 emissions (metric tons)')
plt.xlabel('Years')
plt.xticks(rotation='vertical')


plt.show()
```

```python
df_transformed = df_smaller.drop(columns=["Gap"]).set_index('Country Name').T
df_transformed.head(10)
```

```python
df_transformed.plot(kind='line',figsize=(20, 10)
)

plt.ylabel('CO2 emissions (metric tons)')
plt.xlabel('Years')
plt.xticks(rotation='vertical')

plt.legend(
    loc='upper center', 
    bbox_to_anchor=(0.5, 1.25),#position of the legend, first two values are x and y coordinates depending on the figure size
    ncol=3, # number of columns in legend
)


plt.show()
```

```python
plot = sns.boxplot(
    data=df_transformed,
)
plot.tick_params(axis='x', labelrotation = 80)

# Show the plot 


```

```python
plt.figure(figsize=(15,8))
s = sns.lineplot(
    data=df_transformed,
    palette="tab10",
    linewidth=2.5,
    

)
s.tick_params(axis='x', labelrotation = 80)

plt.show()
```

```python
#remove all the columns that are not in the array of the years 1980 to 2018 and recalculate the gap
years = [str(i) for i in range(1980, 2019)]
df_less_year = df_sorted[["Country Name"] + years]
df_less_year['Gap'] = df_less_year.loc[:, '1985':'2018'].max(axis=1) - df_less_year.loc[:, '1985':'2018'].min(axis=1)
df_less_year = df_less_year.sort_values(by='Gap', ascending=False)
df_less_year

```

```python
df_less_year = df_less_year.drop(columns=["Gap"])

df_less_year_gap = df_less_year.head(20)
df_less_year_gap.head(10)
```

```python
df_transformed_less_year = df_less_year_gap.set_index('Country Name').T
df_transformed_less_year.reset_index(inplace=True)
df_transformed_less_year.columns
df_transformed_less_year.rename(columns={'index': 'Year'}, inplace=True)
df_transformed_less_year.set_index("Year", inplace=True)

df_transformed_less_year.head(10)
```

```python
df_transformed_less_year
```

```python
df_transformed_less_year.plot(kind='line',figsize=(20, 10)
)

plt.ylabel('CO2 emissions (metric tons)')
plt.xlabel('Years')
plt.xticks(rotation='vertical')

plt.legend(
    loc='upper center', 
    bbox_to_anchor=(0.5, 1.25),#position of the legend, first two values are x and y coordinates depending on the figure size
    ncol=3, # number of columns in legend
)


plt.show()
```

```python
df_pie = df_smaller.sort_values(by=['1960'],ascending=False).head(10)
df_pie
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(6,6))
plt.pie(df_pie["1960"], autopct='%1.1f%%', labels=df_pie["Country Name"], startangle=140)
plt.title('The largest CO2 emissions in 1960')
plt.show()
```

```python
plt.figure(figsize=(30,3))
plt.bar(df_pie['Country Name'], df_pie['1960'], color='blue',width=1)
plt.ylim(ymin=0)
plt.xlim(xmin=-0.5)
```

```python
import pylab as pl
la = pl.matplotlib.font_manager.FontManager()
lu = pl.matplotlib.font_manager.FontProperties(family = 'Helvetica', weight = 'bold')
la.findfont(lu)
```

```python
font = {'family' : 'Helvetica',
        'weight' : 'bold',
        'size'   : 15}

plt.rc('font', **font)
fig, axes = plt.subplots(int(len(df_transformed_less_year.columns)/4), 4, figsize=(30,20))  # Create a 2x2 grid of subplots
axes = axes.flatten()  # Flatten the axes array for easier iteration


for i, column in enumerate(df_transformed_less_year.columns):
    df_transformed_less_year[column].plot(ax=axes[i], kind='line', title=column)
    axes[i].set_xlabel('Year')

    axes[i].tick_params(axis='x', labelrotation=45)
    
#display the ylabel for the first column
#
    if i % 4 == 0:
        axes[i].set_ylabel('CO2 emissions')
        axes[i].set_ylim([0, 100])
        #set the x-axis limit for each subplot

    else:
        axes[i].set_ylabel('')
    # Set the y-axis limit for all subplots
        axes[i].set_ylim([0, 50])

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
